```python
import pandas as pd

# Leitura do arquivo CSV
data = pd.read_csv('DATASET.CSV', sep=";")

def calcular_imc(massa, altura):
    imc = massa / (altura ** 2)
    return imc

# Interação sobre cada linha do dataset!!!

for index, row in data.iterrows():
    primeiro_nome = row["Primeiro Nome"].upper().strip()
    sobrenomes = row['Sobrenomes'].upper().strip()
    nome_completo = f"{primeiro_nome} {sobrenomes}"

# Verificar se os valores de Peso e Altura são válidos
    massa = row['Peso (kg)']
    altura = row['Altura (m)']

# Verificar se os valores de massa e altura são strings
    if isinstance(massa, str):
        try:
            massa = float(massa.replace(",","."))
        except ValueError:
            print(f"Erro: Valor inválido para a massa em{nome_completo}. Ignorando registro.")
            continue

    if isinstance(altura, str):
        try:
            altura = float(altura.replace(",","."))
        except ValueError:
            print(f"Erro: Valor inválido para a altura em{nome_completo}. Ignorando registro.")
            continue

# Calcular o IMC
    imc = calcular_imc(massa, altura)

# Imprimir nome completo e IMC formatado
    print("{:<25} {:.2f}".format(nome_completo, imc).replace('.' ,','))

```
